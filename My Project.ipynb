```python
# Question 1 

import numpy as np
import pandas as pd 
import csv
import matplotlib.pyplot as plt

file=pd.read_csv('/Users/mukulhooda/Downloads/startup_funding.csv')
df=file.copy()

locations=['Bangalore','Mumbai','Gurgaon','Noida','New Delhi']

df['CityLocation'].fillna(" ",inplace=True)
df = df[df["CityLocation"] != " "]


np_location=np.array(df['CityLocation'])
for i in range(len(np_location)):
    if 'Delhi' in np_location[i]:
        np_location[i]='New Delhi'
    if 'bangalore' in np_location[i]:
        np_location[i]='Bangalore'

dict={}

for i in np_location:
    l=i.split('/')
    for c in l:
        city=c.strip()
        if city in locations:
            if city in dict:
                dict[city]+=1
            else:
                dict[city]=1

Preferable_locations=np.array(list(dict.keys()))
no_of_fundings=np.array(list(dict.values()))

plt.bar(Preferable_locations,no_of_fundings)
plt.ylabel('Number of Fundings',size=17)
plt.xlabel('Preferable Locations',size=17)
plt.title('Location Versus Number of Fundings',size=20)
plt.yticks(size=15)
plt.xticks(rotation=90,size=15)
plt.show()

print('Locations and respective number of times funding recieved:')
for i in range(len(Preferable_locations)):
    print(Preferable_locations[i],' ---> ',no_of_fundings[i])
    
index=no_of_fundings.argmax()
print()
print('Maximum funding is done in',Preferable_locations[index])
```

```python
# Question 2

import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

file=pd.read_csv('/Users/mukulhooda/Downloads/startup_funding.csv')

df=file.copy()
df=df.dropna(subset=['InvestorsName'])
df=df[df.InvestorsName != '']
df = df[df.InvestorsName != 'Undisclosed Investors']
df = df[df.InvestorsName != 'Undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investors']
df = df[df.InvestorsName != 'undisclosed investor']


investors = df.InvestorsName
investor_name = []
for i in investors:
    if i != "":
        temp = i.split(",")
        for j in temp:
            investor_name.append(j.strip())
            
dict = {}
for i in investor_name:
    dict[i] = dict.get(i,0) + 1



x=[]
y=[]
for i,j in dict.items():
    x.append(i)
    y.append(j)
    
np_x=np.array(x)
np_y=np.array(y)

np_x=np_x[np.argsort(np_y)]
np_y=np.sort(np_y)
    
np_y=np_y[::-1]
np_x=np_x[::-1]

    

xaxis=np_x[:5]
yaxis=np_y[:5]

plt.bar(xaxis,yaxis,color='pink')
plt.xticks(rotation=90,size=15)
plt.yticks(size=15)
plt.xlabel('Investors --->',size=17)
plt.ylabel('No. of times Invested --->',size=17)
plt.title('Investors Versus No. of times invested',size=20)
plt.show

print('Top 5 investors:')
for i in range(len(xaxis)):
    print(xaxis[i],'--->',yaxis[i])
print()
print('Top investor is ',xaxis[0])
```

```python
# Question 3

import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

file=pd.read_csv('/Users/mukulhooda/Downloads/startup_funding.csv')

df=file.copy()
df.dropna(subset=['InvestorsName','StartupName'],inplace=True)


np_investor=np.array(df['InvestorsName'])
np_startup=np.array(df['StartupName'])
for i in range(len(np_startup)):
    if 'Ola' in np_startup[i]:
        np_startup[i]='Ola'
    if 'Flipkart' in np_startup[i]:
        np_startup[i]='Flipkart'
    if 'Oyo' in np_startup[i] or 'OYO Rooms' in np_startup[i]:
        np_startup[i]='Oyo'
    if 'Paytm' in np_startup[i]:
        np_startup[i]='Paytm'
        
dict={}
for i in range(len(np_investor)):
    investor = np_investor[i].split(',')
    for invest in investor:
        invest=invest.strip()
        if invest != "":
            if invest in dict:
                s=dict[invest]
                s.add(np_startup[i])
                dict[invest]=s
            else:
                dict[invest]={np_startup[i]}

for key in dict:
    dict[key]=len(dict[key])

investor_name=np.array(list(dict.keys()))
number_of_fundings=np.array(list(dict.values()))

ind=number_of_fundings.argsort()[::-1]
ind=ind[:5]

investor_name=investor_name[ind]
number_of_fundings=number_of_fundings[ind]


    
plt.bar(investor_name,number_of_fundings,color='brown')
plt.ylabel('Number of Fundings--->',size=17)
plt.xlabel('Investors --->',size=17)
plt.title('Investor Versus Number of Fundings',size=20)
plt.yticks(fontsize=15)
plt.xticks(rotation=90,fontsize=15)
plt.show()

print('Top 5 Investors are:')
for i in range(len(investor_name)):
    print(investor_name[i],'-',number_of_fundings[i])
print()
print('Top investor is ',investor_name[0])
```
